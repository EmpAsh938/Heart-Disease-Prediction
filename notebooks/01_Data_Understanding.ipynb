7. # Evaluation Metric Discussion

Since this is a binary classification problem, evaluating the model using only accuracy may not provide a complete picture of performance.

For example, consider a dataset where:

- 90% of patients do not have heart disease.
- 10% of patients do have heart disease.

A model that predicts every patient as healthy would achieve 90% accuracy, despite failing to identify a single patient with heart disease. Such a model would be clinically useless.

Therefore, additional evaluation metrics will be considered.

### Accuracy

Measures the proportion of correctly classified instances.

Useful when class distributions are balanced but insufficient on its own for medical datasets.

### Precision

Precision measures how many patients predicted as having heart disease actually have the disease.

High precision reduces false alarms.

### Recall (Sensitivity)

Recall measures how many patients with heart disease are correctly identified by the model.

In medical diagnosis, recall is often more important than accuracy because failing to detect a patient with heart disease (False Negative) can delay treatment and increase health risks.

### F1 Score

The F1 Score provides a balance between precision and recall, making it useful when both false positives and false negatives are important.

### ROC-AUC Score

The ROC-AUC Score evaluates the model's ability to distinguish between patients with and without heart disease across different classification thresholds.

A higher ROC-AUC indicates better overall discrimination capability.

### Confusion Matrix

The confusion matrix provides a detailed summary of prediction outcomes:

- True Positives (TP)
- True Negatives (TN)
- False Positives (FP)
- False Negatives (FN)

This helps identify the types of errors the model makes.
